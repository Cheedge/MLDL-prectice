### a. Directly K-fold

```python
# Normally method to induce dataset
# make a dataset
import numpy as np
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=1000, random_state=0)

# split train and test dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=1)




# covert list into pandas DataFrame


```

here avoid use the normal split method instead using k-fold method

```python
# directly use K-Fold
# induece split method (cross-valiation method)
from sklearn.model_selection import KFold, StratifiedKFold, StratifiedShuffleSplit
kf = KFold(n_splits=100, random_state=3, shuffle=True)
skf = StratifiedKFold(n_splits=100, random_state=2, shuffle=True)
sss = StratifiedShuffleSplit(n_splits=100, random_state=4)


# loop over folds:
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, fbeta_score, roc_auc_score
for train_index, test_index in kf.split(X, y):
#     print(type(train_index))
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    # use scaler
    sc = MinMaxScaler()
    sc_data = sc.fit(X_train)
    X_train_sc = sc.transform(X_train)
    X_test_sc = sc.transform(X_test)
    
    # define a classifier
    clf=DecisionTreeClassifier(max_depth=3,random_state=12)

    model=clf.fit(X_train_sc,y_train)

    pred=model.predict(X_test_sc)
    print ("recall : ",recall_score(pred,y_test))

    
```

### b. Hyperparameter + Pipline + K-fold

```python
# explicitly require this experimental feature
from sklearn.experimental import enable_halving_search_cv  # noqa
# now you can import normally from model_selection
from sklearn.model_selection import HalvingGridSearchCV
from sklearn.model_selection import HalvingRandomSearchCV
from sklearn.ensemble import RandomForestClassifier

# 1, set params:
param_grid = {'model__max_depth': [3, 5, 10],
             'model__min_samples_split': [2, 5, 10]}
# 2, choose used estimator
choose_estimator = RandomForestClassifier(random_state=0)
print(choose_estimator.get_params().keys())

# 3, joint use of Pipline
#Using a pipeline now to replicate the same process we did above 
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
pipeline = Pipeline([
 ('scaling', MinMaxScaler()),
 ('model',choose_estimator)
])
print('\n', pipeline.get_params().keys())

# 4, use ensemble
# ensemble = HalvingGridSearchCV(choose_estimator, param_grid, cv=kf, factor=2, resource='n_estimators',
#                          max_resources=30)
ensemble = HalvingGridSearchCV(pipeline, param_grid,scoring="accuracy", factor=100,
                               cv=sss,return_train_score=True, n_jobs=-1)
ensemble.fit(X_train, y_train)


print ("\n\nGrid Search Results\nScores for 5 splits : ")
print( ensemble.cv_results_)
# print( ensemble.cv_results_['split0_test_score'])
# print( ensemble.cv_results_['split1_test_score'])
```

```python
print(ensemble.best_estimator_, ensemble.best_index_, ensemble.best_params_, ensemble.best_score_)
```

## Example (use Taitanic Dataset)

```python
# Use Taitanic Datasets
import pandas as pd
train_df = pd.read_csv('./Datas/train.csv') #, 'r', error_bad_lines=False)
test_df = pd.read_csv('./Datas/test.csv') #, 'r', error_bad_lines=False)


# 0. Preprocessing data: encoding, and NaN data
from sklearn.preprocessing import LabelEncoder
encode = LabelEncoder()
train_df['Sex'] = encode.fit_transform(train_df['Sex'])
test_df['Sex'] = encode.fit_transform(test_df['Sex'])
train_df['Age'] = train_df['Age'].fillna(train_df['Age'].median())
test_df['Age'] = test_df['Age'].fillna(test_df['Age'].median())
train_df['Fare'] = train_df['Fare'].fillna(train_df['Fare'].median())
test_df['Fare'] = test_df['Fare'].fillna(test_df['Fare'].median())
# convert Pclass feature to 3
train_df = pd.get_dummies(train_df, columns=['Pclass'])
test_df = pd.get_dummies(test_df, columns=['Pclass'])
test_df.columns

features = ['Sex', 'Age', 'Fare', 'Pclass_1', 'Pclass_2', 'Pclass_3']
y_train = train_df['Survived']
X_test = test_df[features]
X_train = train_df[features]
train_df

# 0. make train validation datasets
# features0 = ['Sex', 'Age', 'Fare', 'Pclass_1', 'Pclass_2', 'Pclass_3', 'Survived']
df_X = train_df[features]
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(train_df[features], train_df['Survived'],
                                                     test_size=0.2, random_state=1)


# 1. choose estimator
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
dt_clf = DecisionTreeClassifier(random_state=0)
svm_clf = SVC(random_state=0)
knn_clf = KNeighborsClassifier(n_jobs=-1)
rf_clf = RandomForestClassifier(random_state=0, n_jobs=-1)
gb_clf = GradientBoostingClassifier(random_state=0)

choose_clf = rf_clf
print(choose_clf.get_params().keys())

# 2. pipeline
from sklearn.pipeline import Pipeline
pipeline = Pipeline([('estimator', choose_clf)])
# print(pipeline.get_params().keys())

# 3. set params
dt_param = {'estimator__criterion': ['entropy', 'gini'],
            'estimator__max_depth': [3, 10],
            'estimator__max_features': [10]}
svm_param = {'estimator__krnel': ['linear', 'rbf', 'sigmoid'],
             'estimator__C':[0.025]}
knn_param = {'estimator__algorithm': ['ball_tree', 'kd_tree']}
rf_param = {'estimator__max_depth': [3, 10],
            'estimator__max_features': [6],
            'estimator__n_estimators': [100, 300]}
gb_param = dict(estimator__learning_rate = [0.1, 0.01],
                estimator__loss = ['deviance', 'exponential'],
                estimator__criterion = ['friedman_mse', 'mse', 'mae'],
                estimator__n_estimators = [100, 300],
                estimator__max_depth = [3, 5])
choose_param = rf_param

# 4. k-fold (shuffle)
from sklearn.model_selection import KFold, StratifiedKFold, StratifiedShuffleSplit
kf = KFold(n_splits=300, shuffle=True, random_state=0)
skf = StratifiedKFold(n_splits=300, shuffle=True, random_state=0)
sss = StratifiedShuffleSplit(n_splits=300, random_state=0)
choose_cv = kf

# 5. ensemble
from sklearn.experimental import enable_halving_search_cv
from sklearn.model_selection import HalvingGridSearchCV
from sklearn.model_selection import HalvingRandomSearchCV

ensemble_hgs = HalvingGridSearchCV(pipeline, choose_param, factor=100, n_jobs=-1, cv = choose_cv, 
                                  scoring='accuracy', return_train_score=True, min_resources=600)
ensemble_hrs = HalvingRandomSearchCV(pipeline, choose_param, factor=100, n_jobs=-1, cv = choose_cv,
                                     scoring='accuracy', return_train_score=True)
# print(ensemble_hgs.get_params())

```

```python
ensemble_hgs.fit(X_train, y_train)
```

```python
ensemble_hgs.best_estimator_
ensemble_hgs.best_score_
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, fbeta_score, roc_auc_score
print('accuracy={0:0.3f}'.format(accuracy_score(y_valid, ensemble_hgs.predict(X_valid))))
print('precision={0:0.3f}'.format(precision_score(y_valid, ensemble_hgs.predict(X_valid))))
print('recall={0:0.3f}'.format(recall_score(y_valid, ensemble_hgs.predict(X_valid))))
print('f1={0:0.3f}'.format(f1_score(y_valid, ensemble_hgs.predict(X_valid))))
print('f2={0:0.3f}'.format(fbeta_score(y_valid, ensemble_hgs.predict(X_valid), beta=2)))
print('f0.5={0:0.3f}'.format(fbeta_score(y_valid, ensemble_hgs.predict(X_valid), beta=0.5)))
print('roc_auc={0:0.3f}'.format(roc_auc_score(y_valid, ensemble_hgs.predict(X_valid))))
```
