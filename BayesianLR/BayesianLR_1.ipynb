```python
from BayesianPlotUtensil import *
import matplotlib.pyplot as plt
import numpy as np
#%matplotlib inline

N = 30
x0= np.random.rand(N,1)

# print(x0.shape, x.shape)
w1, w0 = 2, 0.3
variance=0.04

def lin_func(x, w1, w0, variance):
    noise = np.random.normal(scale=np.sqrt(variance), size=x.shape)
    y = w1 * x + noise + w0
    return y

y_noise = lin_func(x0, w1, w0, variance)

x_line0 = np.linspace(0, 1, 100).reshape(-1, 1)
y_true = lin_func(x_line0, w1, w0, 0)

plt.plot(x_line0, y_true, 'b--')
plt.scatter(x0, y_noise, color='green')
```

```python
from BayesianLR import BayesianLinearRegression
# variance=sigma**2
sigma = 0.2
tau = 0.707
# x = np.c_[x0, np.ones(x0.shape)]
# x_line = np.c_[x_line0, np.ones(x_line0.shape)]

BayesLR = BayesianLinearRegression(x0, y_noise, sigma, tau)
theta_mean, theta_cov = BayesLR.Posterior()
# x_test = np.random.rand(1,1)
y_hat_mu, y_hat_std = BayesLR.predict(x_line0)
print(theta_mean.shape, theta_cov.shape, y_hat_mu.shape, y_hat_std.shape)
```

```python
plt.plot(x_line0, y_true, 'k--')
plt.scatter(x0, y_noise, color='green')

PlotBayesianConfidenceRegion(x_line0, y_hat_mu, np.sqrt(y_hat_std))
# print(theta_mean.shape,theta_cov.shape)

# PlotPosteriorFitting(x, theta_mean, theta_cov, 5)
```
